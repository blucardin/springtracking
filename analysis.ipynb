```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('positions1770977671.9285948.csv', delimiter=',')

x_data = data[:,0]
y_data = data[:,1]

length_of_video = 91 # s
time = np.linspace(0, length_of_video, len(y_data))# seconds
```

```python
y_data
```

```python
plt.plot(time, y_data, label="y-position")

plt.plot(time, x_data, label="x-position")
```

```python
left_data_point = 350
right_data_point = 2500
y_data_trimmed = y_data[left_data_point:right_data_point]
time_trimmed = time[left_data_point:right_data_point]
x_data_trimmed = x_data[left_data_point:right_data_point]
```

```python
len(y_data_trimmed)
```

```python


plt.figure(figsize=(11, 3))

plt.plot(time_trimmed, y_data_trimmed, label="y-position")

plt.xlabel("Time (s)")
plt.ylabel("Distance (flipped)")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(11, 3))

plt.plot(time_trimmed, x_data_trimmed, label="x-position")

plt.xlabel("Time (s)")
plt.ylabel("Distance (flipped)")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(11, 3))

plt.plot(time_trimmed, np.gradient(y_data_trimmed), label="derivative")

plt.legend()

plt.xlabel("Time (s)")
plt.ylabel("Distance Over Time (flipped)")

```

```python

yf = np.abs(np.fft.fft(y_data_trimmed))
freq = np.fft.fftfreq(len(y_data_trimmed), time_trimmed[0] - time_trimmed[1])

# freq = freq[:len(freq)//2 + 1]

n = len(yf)

fig, ax = plt.subplots()
ax.plot(freq, yf)
plt.xlim((0, 3))
plt.ylim((0, 200000))
plt.xlabel("Frequency (Cycles/Second)")
plt.ylabel("Amplitude (I think)")

plt.show()

```
